```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score, roc_auc_score,roc_curve
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression

%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
pd.set_option('display.max_rows',None)
```

```python
#Getting the first letter of the cabin
train['Deck'] = train['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'M')
train.drop(['Cabin'],axis=1,inplace=True)
```

```python
#Getting the first letter of the cabin
test['Deck'] = test['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'M')
test.drop(['Cabin'], axis=1,inplace=True)
```

```python
train[['Survived','Deck']].groupby('Deck')['Survived'].mean().plot(kind='bar',figsize=(15,7))
```

```python
df_all = pd.concat([train,test],axis=0)
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',hue='Sex',data=df_all)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    Sex = cols[2]
    if pd.isnull(Age):
        if Pclass == 1:
            if Sex == 'male':
                return 42
            else:
                return 36
        elif Pclass == 2:
            if Sex == 'male':
                return 30
            else:
                return 28
        else:
            if Sex == 'male':
                return 24
            else:
                return 22
    else:
        return Age
train['Age'] = train[['Age','Pclass','Sex']].apply(impute_age,axis=1)
test['Age'] = train[['Age','Pclass','Sex']].apply(impute_age,axis=1)
```

```python
train[['Deck','Embarked']].groupby('Deck')['Embarked'].value_counts()
```

```python
train.loc[train['Embarked'].isnull()]
```

```python
train.sort_values(by='Fare')
```

```python
train['Embarked'].fillna(method='ffill', inplace=True)
test['Embarked'].fillna(method='ffill', inplace=True)
```

```python
test.isnull().sum()
```

```python
test[['Fare','Pclass']].groupby('Pclass')['Fare'].mean()
test.loc[test['Fare'].isnull()]
```

```python
test['Fare'].fillna(12.459678, inplace=True)
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
sex.rename({'male':'Male'},axis=1,inplace=True)
```

```python
sex_test = pd.get_dummies(test['Sex'],drop_first=True)
sex_test.rename({'male':'Male'},axis=1,inplace=True)
```

```python
embarked = pd.get_dummies(train['Embarked'],drop_first=True)
embarked_test = pd.get_dummies(test['Embarked'],drop_first=True)
deck = pd.get_dummies(train['Deck'],drop_first=True)
deck_test = pd.get_dummies(test['Deck'],drop_first=True)
```

```python
train.drop(['Sex','Embarked','Deck'],axis=1,inplace=True)
test.drop(['Sex','Embarked','Deck'],axis=1,inplace=True)
```

```python
train=pd.concat([train,embarked,deck,sex],axis=1)
test=pd.concat([test,sex_test,embarked_test,deck_test],axis=1)
```

```python
# Get Title from Name
title = [i.split(",")[1].split(".")[0].strip() for i in train["Name"]]
title = pd.Series(title)
title.value_counts()
```

```python
title.replace(['Rev','Col','Major','Sir','Lady','the Countess','Jonkheer','Capt','Don'],'Rare',inplace=True)
```

```python
title.replace({'Mlle':'Miss',
               'Ms':'Miss',
               'Mme':'Miss'},inplace=True)
```

```python
title.value_counts()
```

```python
train.drop(['T'],axis=1,inplace=True)
```

```python
title_test = [i.split(",")[1].split(".")[0].strip() for i in test["Name"]]
title_test = pd.Series(title_test)
title_test.replace(['Rev','Col','Major','Sir','Lady','the Countess','Jonkheer','Capt','Don','Dona'],'Rare',inplace=True)
title_test.replace({'Mlle':'Miss',
               'Ms':'Miss',
               'Mme':'Miss'},inplace=True)
```

```python
title_test.value_counts()
```

```python
title = pd.get_dummies(title, drop_first=True)
title_test = pd.get_dummies(title_test, drop_first=True)
```

```python
train = pd.concat([train,title],axis=1)
test = pd.concat([test,title_test],axis=1)
```

```python
X_1 = train.drop(['Name','Ticket','Survived','PassengerId'],axis=1).copy()
y_1 = train['Survived'].copy()
```

```python
X_1.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_1,y_1,test_size=0.25, 
                                                    random_state=0)
```

```python
model_1= RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model_1.fit(X_train,y_train)
pred_1 = model_1.predict(X_test)
```

```python
acc_score_1= accuracy_score(y_test,pred_1)
acc_score_1
```

```python
print(classification_report(y_test,pred_1))
```

```python
model_2 = XGBClassifier(n_estimators = 500, random_state=0,learning_rate=0.5)
model_2.fit(X_train,y_train)
pred_2 = model_2.predict(X_test)
acc_score_2 = accuracy_score(y_test,pred_2)
acc_score_2
```

```python
print(classification_report(y_test,pred_2))
```

```python
model_3 = LogisticRegression()
model_3.fit(X_train,y_train)
pred_3 = model_3.predict(X_test)
acc_score_3 = accuracy_score(y_test,pred_3)
acc_score_3
```

```python
print(classification_report(y_test,pred_3))
```
